```python
# 줄바꿈없이 다음 코드를 출력(힌트: end='') print 함수는 두 번 사용합니다.세미콜론 (;)은 한줄에 여러 개의 명령을 작성하기 위해 사용합니다.
print("first", end= "");print("second")
```
