# 082 별표현식 2
다음과 같이 10개의 값이 저장된 scores 리스트가 있을 때, start expression을 사용하여 
우측 8개의 값을 valid_score 변수에 바인딩하여라.


```python

scores = [8.8, 8.9, 8.7, 9.2, 9.3, 9.7, 9.9, 9.5, 7.8, 9.4]
*vaild_score, _, _= scores
print(vaild_score)
```

# 083 별표현식 3
다음과 같이 10개의 값이 저장된 scores 리스트가 있을 때, 
start expression을 사용하여 가운데 있는 8개의 값을 valid_score 변수에 바인딩하여라.


```python

scores = [8.8, 8.9, 8.7, 9,2, 9.3, 9.7, 9.9, 9.5, 7.8, 9.4]
_, *vaild_score2, _ = scores
print(vaild_score2)
```
